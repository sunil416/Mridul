```python
folder = "/content/sample_data/"
documents = SimpleDirectoryReader(folder, required_exts=[".json"])

docs=documents.load_data()
```

```python
from llama_index.core import Settings
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
```

```python
llm = GoogleGenAI(model="gemini-2.5-flash",api_key=GoogleAPI)  # Note updated model naming
embed_model = GoogleGenAIEmbedding(model_name='models/gemini-embedding-exp-03-07',api_key=GoogleAPI)

# Configure Settings to use these globally in LlamaIndex
Settings.llm = llm
Settings.embed_model = embed_model
index = VectorStoreIndex.from_documents(
    docs,
    llm=llm,
    embed_model=embed_model
)
```
